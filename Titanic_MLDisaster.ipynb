```python
# Exploring Age feature missing values for imputation
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import scipy.stats as stats

# Set up the matplotlib figure
plt.figure(figsize=(8, 6))

# Plot the distribution of Age to recognize if age feature is normally distributed
for i in ['Age','Fare']:
  sns.histplot(data_train[i].dropna(), kde=True, bins=40)
  plt.title("{0} Distribution".format(i))
  plt.xlabel("{0}".format(i))
  plt.ylabel('Frequency')
  plt.show()
```

### Age and Fare distribtions are more skewed than normally distributed, therefore, using median function for handling these missing values is preferable

```python
# Fill missing ages with median value since it skewed distributed
data_train['Age'].fillna(data_train['Age'].median(),inplace=True)
data_test['Age'].fillna(data_test['Age'].median(),inplace=True)
data_test['Fare'].fillna(data_test['Fare'].median(),inplace=True)
```

```python
# Fill missing cabins with 'Unknown',
# also should be separted into characters and numbers
# there are also more than one string so maybe it means there are more than one cabin to one passangerID
data_train['Cabin'].fillna('Unknown', inplace=True)
data_test['Cabin'].fillna('Unknown', inplace=True)
```

```python
# Fill missing embarked values with the first mode
data_train['Embarked'].fillna(data_train['Embarked'].mode()[0], inplace=True)
data_test['Embarked'].fillna(data_test['Embarked'].mode()[0], inplace=True)
```

```python
# chacking fillings of missings values both on training and testing data
print(data_train.isnull().sum())
print(data_test.isnull().sum())
```

# Data Preprocessing & Feature Engineering - DONE!
No significant feature engineering required

```python
# creating a copy of the data to preserve our original data
dtrain = data_train.copy()
dtest = data_test.copy()
```

```python
# encoder sex feature as a binary
from sklearn.preprocessing import LabelEncoder

LE_sex = LabelEncoder()

dtrain['Sex'] = LE_sex.fit_transform(dtrain['Sex'])
print(dtrain.Sex.unique())

dtest['Sex'] = LE_sex.fit_transform(dtest['Sex'])
print(dtest.Sex.unique())
```

```python
# Embarked indicates the port where each passenger boarded the ship.
# One-Hot Encoding is also applyable

import pandas as pd
import numpy as np
from sklearn.preprocessing import OneHotEncoder

OHE = OneHotEncoder(categories = 'auto', sparse_output = False)

# Fit and transform the 'Embarked' column
Embarked_Encoded_train = OHE.fit_transform(dtrain[['Embarked']])
Embarked_Encoded_test = OHE.fit_transform(dtest[['Embarked']])

# Get the feature names for the encoded columns
Feature_Names = OHE.get_feature_names_out(['Embarked'])

# Create a DataFrame with the encoded data
Encoded_data_Embarked_train = pd.DataFrame(Embarked_Encoded_train, columns = Feature_Names)
Encoded_data_Embarked_test = pd.DataFrame(Embarked_Encoded_test, columns = Feature_Names)

#Concatenate the encoded columns back to the original DataFrame
dtrain = pd.concat([dtrain, Encoded_data_Embarked_train], axis=1)
dtest = pd.concat([dtest, Encoded_data_Embarked_test], axis=1)

# Drop the original 'Embarked' column if no longer needed
dtrain.drop('Embarked', axis=1, inplace=True)
dtest.drop('Embarked', axis=1, inplace=True)
```

```python
# extract person intro in the name feature
dtrain['Person'] = dtrain['Name'].apply(lambda x: x.split(',')[1].split('.')[0].strip())
dtest['Person'] = dtest['Name'].apply(lambda x: x.split(',')[1].split('.')[0].strip())
```

```python
# Ticket Feature
# For Ticket_String
dtrain['Ticket_String'] = dtrain['Ticket'].apply(lambda x: x[:x.rfind(' ') + 1] if ' ' in x else '')
dtest['Ticket_String'] = dtest['Ticket'].apply(lambda x: x[:x.rfind(' ') + 1] if ' ' in x else '')

# For Ticket_No
dtrain['Ticket_No'] = dtrain['Ticket'].apply(lambda x: x.split(' ')[-1] if ' ' in x else x)
dtest['Ticket_No'] = dtest['Ticket'].apply(lambda x: x.split(' ')[-1] if ' ' in x else x)

```

```python
# Cabin Feature
# creating new column for how many cabins per passanager ID
def counting_spaces(cabin):
  cabin = cabin.split(' ')
  if cabin == 'Unknown':
    return 0
  else:
    return len(cabin)

dtrain['Number_of_Cabins'] = dtrain['Cabin'].apply(counting_spaces)
dtest['Number_of_Cabins'] = dtest['Cabin'].apply(counting_spaces)

```

```python
# Cabin Feature
# creating new column for what category cabin is belong to

def cabin_category(cabin):
  if cabin == 'Unknown':
    return 'Unknown'
  else:
    return cabin[0]

dtrain['Cabin_Category'] = dtrain['Cabin'].apply(cabin_category)
dtest['Cabin_Category'] = dtest['Cabin'].apply(cabin_category)
```

```python
# understanding the extended features after some manipulations
object_list = []
int_list = []

for col in data_train.columns:
    if dtrain[col].dtype == 'object':
        object_list.append(col)
    elif dtrain[col].dtype == 'int' or dtrain[col].dtype == 'float' :
        int_list.append(col)

print("all object features = ", object_list)
print("all numbers features = ", int_list)
print(dtrain.info())
print(dtest.info())
```

```python
# dropping unnecessary features
dtrain = dtrain.drop(['PassengerId','Name','Ticket','Cabin'],axis=1)
dtest = dtest.drop(['PassengerId','Name','Ticket','Cabin'],axis=1)

# person can be categorized (maybe it correlated with sex? and indicates the same?)
# we can sort ticket number by acsending or decreasing order as maybe an indicator first travelers and last trevelers?
```

```python
# Cabin_Category indicates the cabin category for each passenger boarded the ship.
# One-Hot Encoding is also applyable

import pandas as pd
import numpy as np
from sklearn.preprocessing import OneHotEncoder

OHE = OneHotEncoder(categories = 'auto', sparse_output = False)

# Fit and transform the 'Embarked' column
Cabin_Category_train = OHE.fit_transform(dtrain[['Cabin_Category']])
Cabin_Category_test = OHE.transform(dtest[['Cabin_Category']])

# Get the feature names for the encoded columns
Feature_Names = OHE.get_feature_names_out(['Cabin_Category'])

# Create a DataFrame with the encoded data
Encoded_Cabin_Category_train = pd.DataFrame(Cabin_Category_train, columns = Feature_Names)
Encoded_Cabin_Category_test = pd.DataFrame(Cabin_Category_test, columns = Feature_Names)

#Concatenate the encoded columns back to the original DataFrame
dtrain = pd.concat([dtrain, Encoded_Cabin_Category_train], axis=1)
dtest = pd.concat([dtest, Encoded_Cabin_Category_test], axis=1)

# Drop the original 'Embarked' column if no longer needed
dtrain.drop('Cabin_Category', axis=1, inplace=True)
dtest.drop('Cabin_Category', axis=1, inplace=True)
```

```python
# Feature Engineering
# applying this when dealing with featuring after cleaning data and encoding
# creating new feature for family size

dtrain['FamilySize'] = dtrain['SibSp'] + dtrain['Parch'] + 1
dtrain = dtrain.drop(['SibSp','Parch'], axis=1)

dtest['FamilySize'] = dtest['SibSp'] + dtest['Parch'] + 1
dtest = dtest.drop(['SibSp','Parch'], axis=1)
```

```python
dtrain = dtrain.drop(['Person','Ticket_String',	'Ticket_No'],axis=1)
dtest = dtest.drop(['Person','Ticket_String',	'Ticket_No'],axis=1)
```
